```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
categories = ['Medusa R64, T3', 'R32, T3', 'R60, T5', 'R128, T5', 'R256, T10', 'R512, T10']
field1 = [30, 30.17945274707744, 29.838223490971167, 31.207702288681798, 38.90627674111024, 71.50174109970666]
field2 = [9.423630151588725, 5.819384556632649, 8.789019162927836, 8.853575317096245, 9.009107568203344, 16.684447239342198]
field3 = [0, 4.92876699413725, 7.4081151820496896, 7.4168571374434595, 7.427099171133495, 13.610598966621579]
field4 = [1.2350743119444068, 2.355523350695703, 3.055198387936671, 3.7907612398387713, 9.619354486350995, 15.961703732088067]

# Set the figure size
plt.figure(figsize=(10, 6))

# Create the stacked bar plot
plt.bar(categories, field1, label='base')
plt.bar(categories, field2, bottom=field1, label='ea_layer')
plt.bar(categories, field3, bottom=np.array(field1) + np.array(field2), label='lm_head')
plt.bar(categories, field4, bottom=np.array(field1) + np.array(field2) + np.array(field3), label='others')

# Customize the plot
plt.ylabel('Total Value')
plt.legend()

# Add total values on top of each bar
for i, category in enumerate(categories):
    total = field1[i] + field2[i] + field3[i] + field4[i]
    plt.text(i, total, f'{total}', ha='center', va='bottom')

# Adjust layout and display the plot
plt.tight_layout()
plt.show()
```
